```python
import numpy as np
import matplotlib.pyplot as plt

# 물리 상수
g = -9.81  # m/s^2

# 초기 조건
y = 100.0  # 높이 (m)
v = 0.0    # 속도 (m/s)

# 시뮬레이션 설정
dt = 0.01
t = 0.0

times = []
heights = []
velocities = []

while y > 0:

    # Euler Method
    v = v + g * dt
    y = y + v * dt

    times.append(t)
    heights.append(y)
    velocities.append(v)

    t += dt

print(f"Ground hit at t = {t:.2f} s")

plt.plot(times, heights)
plt.xlabel("Time (s)")
plt.ylabel("Height (m)")
plt.title("Falling Ball")
plt.grid()
plt.show()
```
