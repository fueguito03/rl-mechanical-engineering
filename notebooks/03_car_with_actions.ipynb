```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 1000.0        # kg
x = 0.0           # m
v = 0.0           # m/s
dt = 0.1          # s

force_scale = 1000.0  # N

actions = [1, 1, 1, 0, -1, -1]
```

```python
times = []
positions = []
velocities = []
acceleration = []
forces = []
```

```python
for step, action in enumerate(actions):
    F = action * force_scale
    a = F / m
    v = v + a * dt
    x = x + v * dt

    times.append(step*dt)
    forces.append(F)
    acceleration.append(a)
    velocities.append(v)
    positions.append(x)
```

```python
plt.plot(times, positions)
plt.xlabel("Times (s)")
plt.ylabel("Positions (m)")
plt.title("x-t graph")
plt.grid()
plt.show()
```

```python
plt.plot(times, velocities)
plt.xlabel("Times (s)")
plt.ylabel("Velocity (m/s)")
plt.title("v-t graph")
plt.grid()
plt.show()
```

```python
plt.plot(times, acceleration)
plt.xlabel("Times (s)")
plt.ylabel("Acceleration (m/s^2)")
plt.title("a-t graph")
plt.grid()
plt.show()
```

```python
plt.plot(times, forces)
plt.xlabel("Times (s)")
plt.ylabel("Force (N)")
plt.title("F-t graph")
plt.grid()
plt.show()
```
