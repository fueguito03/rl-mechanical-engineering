```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class CarEnv:
    def __init__(self):
        self.m = 1000.0
        self.dt = 0.1
        self.force_scale = 1000.0
        self.target = 10.0
        self.x = 0.0
        self.v = 0.0

    def reset(self):
        self.x = 0.0
        self.v = 0.0
        return np.array([self.x, self.v])
    
    def step(self, action):
        F = action * self.force_scale
        a = F / self.m
        self.v = self.v + a * self.dt
        self.x = self.x + self.v * self.dt

        next_state = np.array([self.x, self.v])
        reward = -abs(self.target - self.x)
        done = self.x >= self.target

        return next_state, reward, done
```

```python
env = CarEnv()
state = env.reset()
```

```python
times = []
positions = []
velocities = []
actions = []
rewards = []
```

```python
for step in range(200):
    action = 0
    if env.x < env.target:
        action = 1

    if env.x >= env.target:
        action = -1
    
    next_state, reward, done = env.step(action)
    times.append(step * env.dt)
    positions.append(next_state[0])
    velocities.append(next_state[1])
    actions.append(action)
    rewards.append(reward)

    if done:
        pass
```

```python
plt.plot(times, positions)
plt.xlabel("Time (t)")
plt.ylabel("Positions (m)")
plt.title("Positions - t graph")
plt.grid()
plt.show()
print(positions)
print(done)
```

```python
plt.plot(times, velocities)
plt.xlabel("Time (t)")
plt.ylabel("Velocity (m/s)")
plt.title("Velocity - t graph")
plt.grid()
plt.show()
print(velocities)
```

```python
plt.plot(times, actions)
plt.xlabel("Time (t)")
plt.ylabel("Action ([-1, 0, 1])")
plt.title("Action - t graph")
plt.grid()
plt.show()
print(actions)
```

```python
plt.plot(times, rewards)
plt.xlabel("Time (t)")
plt.ylabel("Rewards")
plt.title("Rewards - t graph")
plt.grid()
plt.show()
print(rewards)
```
