```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 1000 #[kg]
F = 2000 #[N]
x = 0 #[m]
v = 0 #[m/s]
dt = 0.1 #[s]
#총 10초 시뮬레이션
```

```python
times = []
displacement_x = []
velocity = []
```

```python
for step in range(100):
    
    a = F / m
    v = v + a * dt
    x = x + v * dt

    times.append(step * dt)
    velocity.append(v)
    displacement_x.append(x)
```

```python
plt.plot(times, displacement_x)
plt.xlabel("Time (s)")
plt.ylabel("Position (m)")
plt.title("Car Position")
plt.grid()
plt.show()
```

```python
plt.plot(times, velocity)
plt.xlabel("Time (s)")
plt.ylabel("Velocity (m/s)")
plt.title("Car Velocity")
plt.grid()
plt.show()
```
