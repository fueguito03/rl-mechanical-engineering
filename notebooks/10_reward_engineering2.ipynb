```python
def reward_A(prev_distance, new_distance, done):
    # 거리 기반
    reward = -new_distance

def reward_B(prev_distance, new_distance, done):
    # 가까워졌으면 +1, 멀어졌으면 -1
    if prev_distance > new_distance:
        reward = 1
    else:
        reward = -1

    return reward

def reward_C(prev_distance, new_distance, done):
    # 성공하면 +100, 아니면 -1
    if done:
        reward = 100

    else:
        reward = -1

def reward_D(prev_distance, new_distance, done):
    # A + B + C 섞은 형태
    reward = -new_distance

    if new_distance < prev_distance:
        reward += 1
    else:
        reward -= 1
    
    return reward
```
